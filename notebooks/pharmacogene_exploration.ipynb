```python
import pandas as pd
import matplotlib.pyplot as plt

genes = ['CYP2D6', 'CYP2C19', 'TPMT', 'DPYD', 'SLCO1B1']
drugs = [50, 30, 20, 15, 25]

df = pd.DataFrame({
    'Gene': genes,
    'Associated_Drugs': drugs
})

print(df)
```

```python
plt.figure(figsize=(6,4))
plt.bar(df['Gene'], df['Associated_Drugs'])
plt.xlabel('Pharmacogene')
plt.ylabel('Number of Associated Drugs')
plt.title('Drug Associations of Key Pharmacogenes')
plt.xticks(rotation=45)
plt.show()
```
